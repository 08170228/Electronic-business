```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
rs = pd.read_excel('erro.xlsx')
```

```python
rs
```

```python
sns.set(rc = {'figure.figsize':(18,12)})
sns.heatmap(rs.corr(),cmap='coolwarm',annot=True)
```

```python
rs1 = rs[rs["erro"]==1]
rs0 = rs[rs["erro"]==0]
```

```python
Sp1 = rs0.sample(n=9986,axis=0)
a = pd.concat([rs1,Sp1])
a
```

```python
sns.set(rc = {'figure.figsize':(18,12)})
sns.heatmap(a.corr(),cmap='coolwarm',annot=True)
```

```python
X = a.iloc[:, :-1].values
y = a.iloc[:, -1].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
```

```python
print(X_test, X_train, y_test, y_train)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

# Linear

```python
from sklearn.svm import SVC
classifier = SVC(kernel='linear', random_state=0)
classifier.fit(X_train, y_train)

from sklearn.metrics import confusion_matrix, accuracy_score, ConfusionMatrixDisplay
y_pred = classifier.predict(X_test)
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)
print("Accuracy: {:.2f} %".format(accuracy_score(y_test, y_pred)*100))
disp = ConfusionMatrixDisplay(confusion_matrix = cm)
disp.plot()
plt.show()

from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator=classifier, X = sc.fit_transform(X), y = y, cv = 10)
print(accuracies)
print("Accuracy: {:.2f} %".format(accuracies.mean()*100))
print("Standard Deviation: {:.2f} %".format(accuracies.std()*100))
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 0)
classifier.fit(X_train, y_train)

y_pred = classifier.predict(X_test)
cm = confusion_matrix(y_test, y_pred)
print(cm)
print("Accuracy: {:.2f} %".format(accuracy_score(y_test, y_pred)*100))
disp = ConfusionMatrixDisplay(confusion_matrix = cm)
disp.plot()
plt.show()

accuracies = cross_val_score(estimator=classifier, X = sc.fit_transform(X), y = y, cv = 10)
print(accuracies)
print("Accuracy: {:.2f} %".format(accuracies.mean()*100))
print("Standard Deviation: {:.2f} %".format(accuracies.std()*100))
```
